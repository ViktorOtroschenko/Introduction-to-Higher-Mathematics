## Задание №3
1. Напишите код, который будет переводить полярные координаты в декартовы.

2. Напишите код, который будет рисовать график окружности в полярных координатах.

3. Напишите код, который будет рисовать график отрезка прямой линии вполярных координатах.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def polar(R, a):
    x = R * np.cos(a)
    y = R * np.sin(a)
    return x, y
```

```python
z = np.arange(-5 * np.pi, 5 * np.pi, 0.02)
r = 2
x = r * np.cos(z)
y = r * np.sin(z)

plt.polar(x, y)
plt.show()
```

```python
x = np.linspace(-5*np.pi, 5*np.pi, 200)

k = 3
a = 6
b = 4

y = k*x + b
plt.polar(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

2. Решите систему уравнений и неравенств:$$\left \{

y = x^2 – 1 \\
exp(x) + x∙(1 – y) &gt; 1
\end{array}
\right .$$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-5, 5, 200)
y = x**2 - 1

plt.plot(x, y, label=f'$y = x^2 – 1$')

y = (((np.exp(x)) - 1) / x) + 1
plt.plot(x, y, label=f'$exp(x) + x∙(1 – y) = 1$')

plt.legend()
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = np.linspace(-4, 5, 200)

y = np.exp(x) - 1 + x*(2-x**2)
plt.plot(x, y)
plt.grid(True)
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
