Даны два вектора в трехмерном пространстве: (10,10,10) и (0,0,-10)

2) Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
import numpy as np
import math
```

```python
a = np.array([10, 10, 10])
b = np.array([0, 0, -10])
```

## Задание №2
Почему прямые не кажутся перпендикулярными? (см.ролик)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
x = np.linspace(-5, 5, 21)
y = 3 * x + 1
y2 = -(1/3) * x + 1
plt.plot(x, y)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### Ответ: 
Из-за разного масштаба осей.

```python
x = np.linspace(-5, 5, 21)
y = 3 * x + 1
y2 = -(1/3) * x + 1

plt.plot(x, y)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.show()
```

## Задание №3
Напишите код на Python, реализующий построение графиков:
1. окружности,
2. эллипса,
3. гиперболы.

```python
x = np.linspace(-10, 10, 100)
R = 10
y = np.sqrt(R**2 - x**2)
y2 = -np.sqrt(R**2 - x**2)

plt.plot(x, y)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.show()
```

```python
z = np.arange(0, np.pi * 2.0, 0.01)
r = 10
x = r * np.cos(z)
y = r * np.sin(z)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.show()
```

```python
x = np.linspace(-10, 10, 150)
a = 10
b = 3
y = np.sqrt((1-x**2/a**2)*b**2)
y2 = -np.sqrt((1-x**2/a**2)*b**2)

plt.plot(x, y)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.show()
```

## Задание №5

1) Нарисуйте трехмерный график двух параллельных плоскостей.

2) Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# 1.

x = np.linspace(-3,3,10)
y = np.linspace(-3,3,10)
X,Y = np.meshgrid(x,y)
Z= 4*X + 8*Y + 1

X1,Y1 = np.meshgrid(x,y)
Z1= 4*X1 + 8*Y1 + 10

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, Z)
ax.plot_surface(X1, Y1, Z1)
plt.show()
```

```python
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)

x = 10 * np.outer(np.cos(u), np.sin(v))
y = 5 * np.outer(np.sin(u), np.sin(v))
z = 3 * np.outer(np.ones(np.size(u)), np.cos(v))

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(x, y, z)
plt.show()
```
